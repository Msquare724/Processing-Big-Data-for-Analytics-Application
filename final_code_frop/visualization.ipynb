```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
from numpy import genfromtxt
my_data = genfromtxt('new_nyc_temp_crash_edit(1).csv', delimiter=',')

temp = my_data[:, [2]]
crash_numbers = my_data[:, [3]]
```

```python
plt.scatter(crash_numbers, temp, marker='o');
plt.title('NYC crash and temperature scatter graph')
plt.xlabel("Crash Numbers")
plt.ylabel("Temperature(F)");
```

```python
data = np.concatenate((crash_numbers, temp),axis=1)

condition = data[:,0] > 60
big_crash = data[:,:][condition]

plt.scatter(big_crash[:,0], big_crash[:,1], marker='o');
plt.title('NYC crash High and temperature scatter graph')
plt.xlabel("Crash Numbers")
plt.ylabel("Temperature(F)");
```

```python
condition2 = data[:,0] < 10
small_crash = data[:,:][condition2]

plt.scatter(small_crash[:,0], small_crash[:,1], marker='o');
plt.title('NYC low crash and temperature scatter graph')
plt.xlabel("Crash Numbers")
plt.ylabel("Temperature(F)");
```

```python
condition3 = data[:,1] > 95
high_temp = data[:,:][condition3]

plt.scatter(high_temp[:,0], high_temp[:,1], marker='o');
plt.title('NYC low crash and high temperature scatter graph')
plt.xlabel("Crash Numbers")
plt.ylabel("Temperature(F)");
```

```python
condition4 = data[:,1] < 55
low_temp = data[:,:][condition4]

plt.scatter(low_temp[:,0], low_temp[:,1], marker='o');
plt.title('NYC low crash and high temperature scatter graph')
plt.xlabel("Crash Numbers")
plt.ylabel("Temperature(F)");
```

```python
time = my_data[:, [1]]

plt.scatter(time, crash_numbers, marker='o');
plt.title('NYC crash and time scatter graph')
plt.xlabel("time(hour)")
plt.ylabel("Crash Numbers(F)");
```

```python
time_data = np.concatenate((time, crash_numbers),axis=1)

time_data = time_data[time_data[:, 0].argsort()]

means = {}
for i in np.unique(time_data[:,0]):
    tmp = time_data[np.where(time_data[:,0] == i)]
    means[i] = np.mean(tmp[:,1])
```

```python
plt.plot(*zip(*sorted(means.items())))
plt.title('NYC number of average crashes and time')
plt.xlabel("time(hour)")
plt.ylabel("Crash Numbers(F)");
plt.show()
```
